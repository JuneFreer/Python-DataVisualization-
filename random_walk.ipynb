```python
from random import choice
class RandomWalk():
    """一个生成随机漫步数据的类"""
    def __init__(self, num_points=5000): #self指由RandomWalk()类new的对象/实例
        """初始化随机漫步对象的属性"""
        self.num_points = num_points #默认步数5000
        
        """所有随机漫步都始于(0,0)"""
        self.x_values = [0]
        self.y_values = [0]
        
    def fill_walk(self):
        """计算随机漫步包含的所有点"""
        
        # 不断漫步，知道列表长度达到默认步数5000
        while len(self.x_values) < self.num_points:
            
            # 决定前进方向以及沿此方向前进的距离
            x_direction = choice([1, -1]) #正向右，左向负(与数轴同理)，用+1 -1 来控制正负/方向
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
            
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
            
            # 拒绝原地踏步
            if x_distance == 0 and y_distance == 0:
                continue
                
            # 求得下一步后，获取下一个点的位置（x,y）：用现在点的位置 + 下一步
            next_x = self.x_values[-1] + x_step  # self.x_values[-1]代表当前对象(self)的x_values[...]列表中的最后一个值
            next_y = self.y_values[-1] + y_step
            
            self.x_values.append(next_x) # append到列表末尾
            self.y_values.append(next_y)
            
# 下面的代码将随机漫步的所有点都绘制出来
import matplotlib.pyplot as plt

rw = RandomWalk() # 实例化RandomWalk()类，创建一个RandomWalk类的实例
rw.fill_walk() # RandomWalk类的实例调用RandomWalk类的fill_walk()方法

plt.scatter(rw.x_values, rw.y_values, s=5)
plt.show()
```
